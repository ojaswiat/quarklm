# Word Embeddings

**Goal:** Convert words into vectors for mathematical interpretation by algorithms.

## Word Embeddings

### 1. What are they?
Word embeddings are words turned into long lists of numbers (vectors). This math-based format transforms human language into a form that computers can actually calculate and process.

### 2. Why do we use them?
They allow computers to understand the meaning of words. By plotting these numbers in a high-dimensional space, words with similar contexts (like "tea" and "coffee") naturally cluster close together, capturing their relationships.

### 3. How do they work?
1. **Initialize:** Every word is given a random set of numbers (a position in space).
2. **Predict:** The system guesses the next word in a sentence based on its neighbors.
3. **Adjust:** If the guess is wrong, the system shifts the numbers slightly. Over billions of examples, related words are pulled closer together.

### 4. The Math
The similarity of this vectors in the space are calculated using cosine similarity:
$$\text{Cosine Similarity} = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \|\mathbf{B}\|}$$
* **$\mathbf{A}$ and $\mathbf{B}$:** The two word vectors in comparision.
* **$\mathbf{A} \cdot \mathbf{B}$ (Dot Product):** This multiplies the corresponding numbers of both vectors and adds them up. **If the numbers trend in the same direction, this value is high.**
* **$\|\mathbf{A}\|$ and $\|\mathbf{B}\|$ (Vector Magnitude/Length):** This calculates the total length of each vector from the center $(0,0)$ of the grid.
* **The Division ($\frac{\text{Dot Product}}{\text{Lengths}}$):** Dividing by the lengths ensures the formula only measures the *direction* the words point, **ignoring how large the numbers are.**

#### Glossary
* **Cluster**: A portion of the N-dimensional space where similar words (vector representation of the words) are grouped together.
* **Space**: The **N-dimensional space on which the vectors are positioned.
* **Vector**: Set of N numbers representing the position of a word in an N-dimensional space.

## Code Example
Converting words to vector representations using Word2Vec model from Gensim library.

**Note:** This is a very simple example. To see how these pieces connect, see the `examples` folder in the root.

```python
from gensim.models import Word2Vec

# 1. Define a tiny corpus (sentences) to train the model
corpus = [
    ["the", "dog", "barks", "at", "the", "cat"],
    ["a", "cat", "is", "a", "friendly", "pet"],
    ["the", "dog", "loves", "to", "play", "fetch"],
    ["my", "cat", "and", "dog", "sleep", "on", "the", "couch"],
    ["the", "pet", "dog", "chased", "the", "pet", "cat"]
]

# 2. Train Word2Vec with explicit 4D space and minimum configurations
model = Word2Vec(
    sentences=corpus, 
    vector_size=4,      # Sets a 4-dimensional embedding space
    window=2,           # Look at 2 words to the left and right - Sliding Window Size
    min_count=1,        # Don't ignore any words
    epochs=1,           # Number of iterations over the corpus - set to 1 for simplicity and speed
    seed=42             # For reproducible random initialization
)

# 3. Retrieve the 4D vector representations
cat_vector = model.wv['cat']
dog_vector = model.wv['dog']

print("4D Vector for 'cat':", cat_vector)
print("4D Vector for 'dog':", dog_vector)
print("=" * 64)

# 4. Calculate Cosine Similarity using the model's built-in function
similarity = model.wv.similarity('cat', 'dog')
print(f"Cosine Similarity between 'cat' and 'dog': {similarity:.4f}")
print("=" * 64)

# 5. Print conclusions about the similarity
print("Conclusion: low score, increase the epochs to improve the model's understanding of word relationships.")
```

## Why Tokenise before Embeddings?

If an AI maps only whole, exact words to vectors, it suffers from two fatal flaws:

1. **Out-of-Vocabulary Errors:** It cannot process new words, variations, or typos (e.g., treating `"dog"` and `"dogs"` as completely unrelated, or failing entirely on `"LLM-ification"`).
2. **Memory Explosion:** Storing distinct high-dimensional vectors for millions of unique whole words requires an impossible amount of computer memory.

### Why We Use Subword Tokenization

It breaks text into flexible fragments (subwords). This allows a small dictionary of ~50,000 tokens to efficiently reconstruct, look up, and process **any word in existence**—even ones the AI has never seen before—while keeping memory usage low.

### A Note on our Word2Vec Example

In the previous example, **Word2Vec did not skip tokenization**. The input corpus was already pre-tokenized by hand because it was passed as a split list of individual words (`["the", "dog", "barks"...]`) rather than a raw, continuous text string. Word2Vec always requires a tokenized input to function.

#### Glossary
* **Vocabulary:** The fixed master list of all unique tokens (words, characters, and subwords) that a specific model is capable of recognizing and processing. **If a token is not in this master list, the model cannot read it or generate it.**
